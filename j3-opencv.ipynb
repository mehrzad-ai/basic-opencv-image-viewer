```python
import cv2
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
img=cv2.imread('img1.jpg')
```

```python
cv2.imshow('ax1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
plt.show()
```

```python
img_cr=img2[45:680,300:970]
```

```python
plt.close()
plt.imshow(img_cr)
plt.show()
```

```python
img_cr.shape
```

```python
size= (100,100)
img_resize=cv2.resize(img_cr,size,cv2.INTER_AREA)
```

```python
plt.imshow(img_resize)
plt.show()
```

```python
img_gray=cv2.cvtColor(img_cr,cv2.COLOR_RGB2GRAY)
```

```python
plt.close()
plt.imshow(img_gray)
plt.show() 

cv2.imshow('ax1',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def ishow(image):
    cv2.imshow('ax1',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
ret , img_tr=cv2.threshold(img_gray,50,255,0)
ishow(img_tr)
print(ret)
```
